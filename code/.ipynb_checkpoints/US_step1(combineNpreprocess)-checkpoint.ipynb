```python
import requests
import json
import datetime
import pandas as pd
import re
import os
import random
import urllib3
import time
import matplotlib
import nltk

from nltk.corpus import stopwords
from wordcloud import WordCloud,STOPWORDS
from datetime import datetime
from tqdm.notebook import tqdm

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
path = '../../data/us/'
```

# 0. read in data

```python
cnn_df = pd.read_json(path + "cnn.json")
print(cnn_df.shape)
cnn_df.head(3)
```

```python
fox_df = pd.read_json(path+ "foxnews.json")
print(fox_df.shape)
fox_df.head(3)
```

```python
pbs_df = pd.read_json(path + "pbs.json")
print(pbs_df.shape)
pbs_df.head(3)
```

## 1. Preprocess

```python
df = pd.concat([cnn_df, fox_df, pbs_df]).reset_index()

df = df[['title', 'date', 'content', 'source']]

df.head()
```

```python
## create onehot coded keywords recors
df['keyword-trump'] = False
df.loc[df.content.str.contains('|'.join(['Trump', 'Donald Trump'])), 'keyword-trump'] = True

df['keyword-republican'] = False
df.loc[df.content.str.contains('Republican'), 'keyword-republican'] = True

df['keyword-biden'] = False
df.loc[df.content.str.contains('|'.join(['Joe Biden', 'Biden'])), 'keyword-biden'] = True

df['keyword-democratic'] = False
df.loc[df.content.str.contains('|'.join(['Democratic', 'Democrat'])), 'keyword-democratic'] = True

df['keyword-election'] = False
df.loc[df.content.str.contains('|'.join(['presidential election', 'election'])), 'keyword-election'] = True

df.head(3)
```

```python
print(df.shape)
df = df[df[['keyword-trump', 'keyword-republican', 'keyword-biden', 'keyword-democratic', 'keyword-election']].any(axis=1)]
print(df.shape)

```

```python
df['content'] = df['content'].str.replace('\W', ' ', regex=True)
df['content'] = df['content'].str.replace(r'\s+', ' ', regex=True)
```

## subset source to the same amount

```python
# Count the number of observations for each source
source_counts = df['source'].value_counts()

# Find the minimum count among the sources
min_count = source_counts.min()
print(min_count)

# Subset the data to include only observations with the minimum count for each source
subset_df = df.groupby('source').apply(lambda x: x.sample(min_count)).reset_index(drop=True)
```

```python

def show_keyword_weekly_count(data):
#     convert 'date' column to datetime type and set as index

    data = data.copy()
    data['date'] = pd.to_datetime(data['date'])
    
    try:
        data.set_index('date', inplace=True)
    except: pass
#     print(data)
    
    
    # resample data by day and count the number of occurrences
    weekly_counts = pd.DataFrame(data.resample('W').sum())
    g = sns.lineplot(data=weekly_counts)
#     , labels=['Trump', "Republican", "Biden", "Democratic", "Election"]
    plt.axvline(x=pd.to_datetime('2020-11-03'), color='r', linestyle='--', label='2020 Presidential Election')

    plt.legend(loc='upper left')
    plt.show(g)
```

```python
show_keyword_weekly_count(subset_df)
show_keyword_weekly_count(df)
# subset_df.head()
# subset_df
```

```python
def show_source_weekly_count(data):
    
    data = data.copy()
    
    try:
        data.set_index('date', inplace=True)
    except: pass

    # 获取“cnn”来源的数据，并按日期计数
    cnn_data = data[data['source'] == 'cnn'].resample('W').count()

    # 获取“fox news”来源的数据，并按日期计数
    fox_data = data[data['source'] == 'foxnews'].resample('W').count()

    # 获取“pbs”来源的数据，并按日期计数
    pbs_data = data[data['source'] == 'pbs'].resample('W').count()

    # 绘制三个来源的折线图
    plt.plot(cnn_data.index, cnn_data['source'], label='CNN')
    plt.plot(fox_data.index, fox_data['source'], label='Fox News')
    plt.plot(pbs_data.index, pbs_data['source'], label='PBS')
    plt.axvline(x=pd.to_datetime('2020-11-03'), color='r', linestyle='--', label='2020 Presidential Election')

    # 添加标题和轴标签
#     plt.title('Weekly Frequency by Source')
    plt.xlabel('Date')
    plt.ylabel('Frequency')

    # 显示图例
    plt.legend()

    # 显示图形
    plt.show()
```

```python
show_source_weekly_count(df)
show_source_weekly_count(subset_df)
```

```python
def show_keyword_source_barplot(df):
    #show the relation between source and keyword

    data = pd.DataFrame(df.groupby('source')[['keyword-trump', 'keyword-republican' , 'keyword-biden', 'keyword-democratic', 'keyword-election']].sum())
    data = data.reset_index()

    # Melt the DataFrame to long format
    data_melt = pd.melt(data, id_vars='source', var_name='keyword')
#     print(data_melt)

    # Group the data by source and keyword, and calculate the count
    data_grouped = data_melt.groupby(['source', 'keyword'])['value'].sum().reset_index()

#     print(data_grouped)
    
    sns.set_style("white")
    ax = sns.histplot(
        data_grouped,
        x='keyword',
        # Use the value variable here to turn histogram counts into weighted
        # values.
        weights='value',
        hue='source',
        multiple='stack',
        # Add white borders to the bars.
        edgecolor='white',
        # Shrink the bars a bit so they don't touch.
        shrink=0.8
    )
    ax.set(xlabel='Keywords',
           ylabel='Total Articles',
           xticklabels=["Biden", "Democratic",  "Election", "Republican", 'Trump'])
    for i in ax.containers:
        ax.bar_label(i, label_type='center')

    # Remove 'Count' ylabel.
    ax.set_ylabel(None)
    plt.show()
    
```

```python
show_keyword_source_barplot(df)
show_keyword_source_barplot(subset_df)
```

```python
def show_keyword_plot(df):
    freq_trump = df['keyword-trump'].value_counts()[True]
    prec_trump = round(100* subset_df['keyword-trump'].value_counts(normalize=True)[True], 2)

    freq_republican = df['keyword-republican'].value_counts()[True]
    prec_republican = round(100* subset_df['keyword-republican'].value_counts(normalize=True)[True], 2)

    freq_biden = df['keyword-biden'].value_counts()[True]
    prec_biden = round(100* subset_df['keyword-biden'].value_counts(normalize=True)[True], 2)

    freq_democratic = df['keyword-democratic'].value_counts()[True]
    prec_democratic = round(100* subset_df['keyword-democratic'].value_counts(normalize=True)[True], 2)

    freq_election = df['keyword-election'].value_counts()[True]
    prec_election = round(100* subset_df['keyword-election'].value_counts(normalize=True)[True], 2)

    print(freq_trump,prec_trump,freq_republican,  freq_biden, freq_democratic, freq_election)
    freq_data = pd.DataFrame({
        'keyword': ['Trump', 'Republican', 'Election', 'Biden', 'Democratic'],
        'count': [freq_trump,freq_republican, freq_election,  freq_biden, freq_democratic],
        'percentage': [prec_trump, prec_republican, prec_election, prec_biden, prec_democratic]
    })

    matplotlib.rc_file_defaults()
    ax1 = sns.set_style(style=None, rc=None )

    fig, ax1 = plt.subplots(figsize=(12,8))

    sns.lineplot(data = freq_data['percentage'], marker='o', sort = False, ax=ax1)

    ax2 = ax1.twinx()

    sns.barplot(data = freq_data, x='keyword', y='count', alpha=0.5, ax=ax2)
```
