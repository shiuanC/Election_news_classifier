```python
Cti_tsai_df = json_normalize(Cti_tsai.news)
Cti_tsai_df['keyword'] = 'tsai'

Cti_han_df = pd.json_normalize(Cti_han.news)
Cti_han_df['keyword'] = 'han'

Cti_kmt_df = pd.json_normalize(Cti_kmt.news)
Cti_kmt_df['keyword'] = 'kmt'

Cti_dpp_df = pd.json_normalize(Cti_dpp.news)
Cti_dpp_df['keyword'] = 'dpp'
```

```python
# preprocess of the cti data


#concat news metadata from all 4 keywords
Cti_df = pd.concat([Cti_tsai_df, Cti_han_df, Cti_kmt_df, Cti_dpp_df], axis=0)
cols = ['id', 'title', 'release_at', 'news_type.name', 'keyword']
Cti_df = Cti_df[cols]
print(Cti_df.shape)

# only left news from 2019/09/01 to 2020/01/18
Cti_df['release_at'] = pd.to_datetime(Cti_df['release_at'])
Cti_df = Cti_df[(Cti_df['release_at'].dt.date>datetime.date(2019, 9, 1)) & (Cti_df['release_at'].dt.date<datetime.date(2020, 1, 18))]
print(Cti_df.shape)

# conver the keyword data to onehot
Cti_df = pd.concat([Cti_df, pd.get_dummies(Cti_df.keyword, prefix='keyword')], axis=1)
Cti_df.drop('keyword', axis=1, inplace=True)
print(Cti_df.shape)


Cti_df = Cti_df.set_index('id')
Cti_df[['keyword_dpp', 'keyword_han', 'keyword_kmt', 'keyword_tsai']] = Cti_df.groupby('id')['keyword_dpp', 'keyword_han', 'keyword_kmt', 'keyword_tsai'].sum()
Cti_df = Cti_df.drop_duplicates(keep='last')
Cti_df = Cti_df.reset_index()

print(Cti_df.shape)
Cti_df.head()
```

## get a news

```python
Cti_df = pd.read_csv(path + 'cti_news_links.csv')
print(Cti_df.shape)
```

```python
def GetResponse(idx, url): 
    try :
        res = requests.get(url)
        return {"id": idx, "res": res.text}
    except:
        return {"id": idx, "res": None}
```

```python
def GetContent(idx, res):
    try:

        soup = BeautifulSoup(res)
        content = " ".join([p.text for p in soup.find('main', attrs={'class':'article__main'}).find_all('p')[:-1]])
        return {"id": idx, "content": content}
    
    except: 
        return {"id": idx, "content": None}
    
```

```python
Cti_df = Cti_df.set_index('id')
Cti_df.head(3)
```

```python
%timeit
start_time = time.time()

# cti_len = len(Cti_df)
contents = []
LenCti = len(Cti_df)
Inv = 16
TreadNum = 16

for start in tqdm(range(0, LenCti, Inv)):
    
    responses=[]
    end = start+Inv if start+Inv <= LenCti else LenCti
    # start the thread pool

    with ThreadPoolExecutor(TreadNum) as executor:

        # submit tasks and collect futures
        futures = [executor.submit(GetResponse, index, row['url']) for index, row in Cti_df[start: end].iterrows()]

        # process task results as they are available
        for future in as_completed(futures):
            responses.append(future.result())

    tmp_df = pd.DataFrame(responses).set_index("id")
    tmp_df = tmp_df.mask(tmp_df.eq('None')).dropna()
    print(f"{len(tmp_df)} news got a request")
    
    
    with ThreadPoolExecutor(TreadNum) as executor:

        futures = [executor.submit(GetContent, idx, row.res) for idx, row in tmp_df.iterrows()]

        for future in as_completed(futures):
            contents.append(future.result())

    print(f"{len(contents)} out of {len(Cti_df)} got a content\n")

end_time = time.time() 
print(end_time - start_time)
```

```python
contents_df = pd.DataFrame(contents).set_index("id")
# contents_df

Cti_df = Cti_df.merge(contents_df, on='id')
print(Cti_df.shape)
Cti_df = Cti_df.mask(Cti_df.eq('None')).dropna()
print(Cti_df.shape)
```

```python
Cti_df.to_csv(path + 'cti_news.csv', index=False)
```
